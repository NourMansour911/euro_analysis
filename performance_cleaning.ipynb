```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('performance.csv')
```

```python
df
```

```python
df.columns = df.iloc[0] 
df = df.drop(0).reset_index(drop=True) 
```

```python
df
```

```python
df['Team'] = df['Team'].astype(str).apply(lambda x: x.split('[')[0])
```

```python
df.rename(columns={'Title(s)': 'Titles',}, inplace=True)
```

```python
df['Years won'] = df['Years won'].astype(str).apply(lambda x: x.replace('[a]', ' (h)').replace(', (', '(').replace(') ', ') ,').replace(" ",""))
df['Years runner-up'] = df['Years runner-up'].astype(str).apply(lambda x: x.replace('[a]', ' (h)').replace(', (', ' (').replace(') ', ') ,').replace(" ",""))
```

```python
df['Titles'] = df['Titles'].astype(str).map(lambda x: x.replace('—', '0'))
df['Runners-up'] = df['Runners-up'].astype(str).map(lambda x: x.replace('—', '0'))
```
